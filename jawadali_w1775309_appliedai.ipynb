### Part B - Compare and evaluate AI techniques.

I will be discussing a few AI techniques that are used in algorithmic trading. These include Neural Networks, Cluster Analysis, RandomForestClassifier and Reinforcement Learning. I have chosen these techniques as they are reliable and commonly used for algorithmic trading. 

__Reinforcement Learning:__

Reinforcement Learning (RL) is a type of machine learning that allows agents to learn from interactions with their environment. The agent receives feedback in the form of rewards or penalties for its actions and uses this feedback to improve its decision-making over time. The key strengths of RL include its ability to learn from trial and error and adapt to changing environments, making it particularly useful for problems with delayed rewards or uncertain optimal solutions.

One of the main limitations of RL is the difficulty in designing an appropriate reward function, which can greatly affect the agent's performance. Additionally, RL can be intensive, requiring a large amount of data and resources to train.

Despite these limitations, RL offers several advantages, including the ability to optimize decision-making in complex and dynamic environments. Applications of RL include training robots for tasks such as grasping and manipulation, training agents to play games at a superhuman level, and optimizing control systems for autonomous vehicles and power systems.

The type of input data required for RL is typically a description of the current state of the environment and the agent's previous actions, while the expected output is a decision or action for the agent to take. To explore a specific technique in-depth, one could investigate Q-Learning, a popular RL method that involves training an agent to learn a function that maps states to actions, known as the Q-function. The Q-function tells the agent which action to take in a given state in order to maximize the expected cumulative reward over time.

One area where this technique is used is in the field of algorithmic trading. In this context, reinforcement learning is utilized to determine the best actions to take in order to maximize profits over time. The learning agent receives rewards in the form of profits, which allows it to determine the most effective strategy for generating significant returns. The input that reinforcement learning receives is the current market data as the input state and the output is the trading actions taken.

__Neural Networks:__

A form of machine learning algorithm known as a neural network is based on the structure and operation of the human brain. They can be used for a variety of applications, including natural language processing and computer vision, and are particularly well suited for jobs that entail pattern recognition, like picture and speech recognition. 

The ability of neural networks to learn from data and enhance performance over time is one of their main advantages. They can also manage enormous datasets and spot intricate patterns that are difficult for humans to recognise. The drawbacks of this technology do exist, though. Neural network training can be computationally demanding and data intensive. They tend to overfit and might be challenging to interpret.

Neural networks can be used in algorithmic trading for a variety of purposes, such as predictive modelling, portfolio optimization, and the creation of trading algorithms that purchase or sell stocks in response to market conditions. Neural networks often require historical data as its input, such as stock prices, market movements, and trade volumes. A forecast, choice, or action is the anticipated outcome.

Recurrent Neural Networks (RNNs) is a technique that could be used for algorithmic trading. These are a particular class of neural network that work well with time series data and other sequential data. They can be used to simulate intricate interactions between inputs and outputs across time and can deal with input sequences of different durations. They can produce trading signals based on previous data or be used to predict future market patterns.

__Cluster analysis:__

Cluster analysis is a method of grouping similar data points together into clusters or groups based on specific characteristics. This technique is commonly used for exploratory data analysis and is a popular method in unsupervised machine learning.

The benefits of cluster analysis include its ability to spot patterns and structure in data as well as its ability to group similar data points that may be difficult for humans to distinguish. It can also be applied to find anomalies or outliers in data.

This method does have certain drawbacks, such as choosing the right number of clusters to employ and how to interpret the outcomes. The procedure can also be impacted by the distance measure and start-up strategy chosen.

Examples of how cluster analysis can be applied to algorithmic trading include market segmentation, risk management, and identifying similar stocks. Market segmentation involves using cluster analysis to identify different segments of the market and create trading strategies for each segment. Risk management involves using cluster analysis to group stocks with similar risk profiles, allowing for more effective risk management. Identifying similar stocks involves grouping stocks with similar performance characteristics, allowing for more effective portfolio diversification.

The type of input data required for cluster analysis is typically numerical data, such as stock prices, trading volumes, and financial ratios. The expected output from cluster analysis is a group of similar data points or clusters.
A specific technique that could be explored in depth for algorithmic trading is k-means clustering. K-means is a popular and widely used clustering technique that involves dividing a dataset into k clusters, where k is specified beforehand. The algorithm assigns each data point to the cluster with the nearest mean and iteratively updates the cluster centroids and assignments until convergence. K-means algorithm is simple to implement and has been widely used in various application domains including finance.

__RandomForestClassifier:__

The RandomForestClassifier is a machine learning algorithm used for classification tasks. It utilizes a combination of decision trees through ensemble learning, where predictions from multiple decision trees are combined. The algorithm creates a forest of decision trees, each trained on a different subset of the data. When a new data point is presented, each tree makes a prediction and the most commonly predicted outcome is selected as the final result. This approach helps to prevent overfitting, a common problem in single decision trees. The scikit-learn library in Python offers an easy-to-use interface for training and evaluating the model. The algorithm's parameters, such as the number of trees in the forest and the maximum depth of each tree, can be adjusted to improve performance. In summary, the RandomForestClassifier is a powerful algorithm that is useful for classification problems. It trains multiple decision trees on random subsets of data and makes predictions based on the majority vote of the decision trees.

Strengths:
•	The Random Forest algorithm is versatile and can be used for both classification and regression problems.
•	It is able to handle large amounts of data and high-dimensional space.
•	Random Forest is less susceptible to overfitting compared to single decision trees.
•	It can be used for feature selection and determining feature importance.
•	Performance can be improved by parallelizing the algorithm.
Weaknesses:

•	The computational cost of the Random Forest algorithm can be high, especially when working with a large number of trees or high-dimensional data.
•	The inner workings of Random Forest can be hard to interpret as it is considered a "black box" model.
•	Small datasets may not be well suited for Random Forest as the variance among individual decision trees can be high.
•	An excessive number of trees can lead to overfitting of the model.
•	The algorithm may not be suitable for datasets with high levels of noise or uncertainty.

## Downloading price data

```python
import yfinance as yf #calls yahoo finance api
import pandas as pd
```

```python
apple = yf.Ticker("AAPL") #initialise ticker class, downloads price history for apple
apple = apple.history(period="max") #downloads all of apple's history
```

```python
del apple["Dividends"] #removes column
del apple["Stock Splits"] #removes column
```

## Setting up Machine Learning

```python
apple["Tomorrow"] = apple["Close"].shift(-1) #creates tomorrow column and shifts tomorrows close price
```

```python
apple["Target"] = (apple["Tomorrow"] > apple["Close"]).astype(int) #if tomorrows prices if greater than close Target turns into 1 
```

```python
apple = apple.loc["2000-01-01":].copy() #removes all data before 2000
```

## Training machine learning model (RandomForestClassifier)

```python
from sklearn.ensemble import RandomForestClassifier 

model = RandomForestClassifier(n_estimators=200, min_samples_split=100, random_state=1) #number of decision trees, number of overfitting

train = apple.iloc[:-100] #100 rows in train set
test = apple.iloc[-100:] #100 rows in test set 

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"]) #trains model to predict target
```

```python
from sklearn.metrics import precision_score #percentage of times when the market goes up when it says it goes up

preds = model.predict(test[predictors]) #generates predictions
preds = pd.Series(preds, index=test.index) #turns array into panda
precision_score(test["Target"], preds) #calculate precision score
```

accuracy of prediction: 45.7%

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot() #plot our predictions
```

## Building a backtesting system

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
#doing everything we did into one function
```

```python
def backtest(data, model, predictors, start=2500, step=250): #trains model with 10yrs of data #trains model with 250 days
    all_predictions = []

    for i in range(start, data.shape[0], step): #function which loops data year by year
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions) #combines into one data frame
```

```python
predictions = backtest(apple, model, predictors)
```

```python
horizons = [2,5,60,250,1000] #rolling means, 2 days, 5 days, 3 months etc
new_predictors = [] #holds new columns

for horizon in horizons:
    rolling_averages = apple.rolling(horizon).mean() #calculates rolling average
    
    ratio_column = f"Close_Ratio_{horizon}" #creates columns
    apple[ratio_column] = apple["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    apple[trend_column] = apple.shift(1).rolling(horizon).sum()["Target"] #will look at past few days and find sum of target
    
    new_predictors+= [ratio_column, trend_column]
```

```python
apple = apple.dropna(subset=apple.columns[apple.columns != "Tomorrow"]) #removes some columns
```

```python

model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1) 
#increase estimators to 200, and decrease sample split
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.6] = 1 #increases chance of price going up
    preds[preds <.6] = 0 #change threshold
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(apple, model, new_predictors) #run backtest again
```
